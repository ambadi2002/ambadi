```python
q1=data['Sale Price'].quantile(0.25)
q1
```

```python
q3=data['Sale Price'].quantile(0.75)
q3
```

```python
iqr=q3-q1
iqr
```

```python
upper_limit=q3+1.5*iqr
lower_limit=q1-1.5*iqr
```

```python
upper_limit
```

```python
lower_limit
```

```python
outlier=[]

for x in data['Sale Price']:
  if (x>upper_limit) or (x<lower_limit):
    outlier.append(x)
```

```python
outlier
```

```python
data['Sale Price']=data['Sale Price'].clip(lower=lower_limit,upper=upper_limit)
```

**extracting information from renovated year(creating new column with exsisting column)**

```python
data['ever renovated']=np.where(data['Renovated Year']==0,"No","Yes")
```

```python
data.head()
```

```python
# years since renovation = year of sale - year of renovation
```

```python
#extracting year from Date House was Sold

data['purchase year']=pd.DatetimeIndex(data['Date House was Sold']).year
```

```python
data.head()
```

```python
#years since renovation

data['years since renovation']=np.where(data['ever renovated']=='Yes',abs(data['purchase year']-data['Renovated Year']),0)
```

```python
data.head()
```

**deleting columns**

```python
data.drop(columns=['Date House was Sold','purchase year','Renovated Year'],inplace=True)
```
