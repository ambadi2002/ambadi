# **ANOVA**

**ONE WAY ANOVA**



*  A researcher recruits 30 students to participate in a study. The students are randomly assigned to use one of  three studying techniques fot the next three weeks to prapare for an exam. At the end of the three weeks ,all of the students takes the same test.



*   group1=[85,86,88,75,78,94,98,79,71,80]

*   group2=[91,92,93,85,87,84,82,88,95,96]

*  group3=[79,78,88,94,92,85,83,85,82,81]  





```python
group1=[85,86,88,75,78,94,98,79,71,80]
group2=[91,92,93,85,87,84,82,88,95,96]
group3=[79,78,88,94,92,85,83,85,82,81]
```

```python
from scipy.stats import f_oneway
```

```python
f_oneway(group1,group2,group3)
```

```python
f_stat,p_value=f_oneway(group1,group2,group3)
print('f_stat:',f_stat)
print('p_value:',p_value)
```

```python
if p_value<0.05:
  print('Reject null hypothesis')
else:
  print('Accept null hypothesis')
```

**CASE STUDY 2**

```python
data1=pd.read_csv("/content/PlantGrowth.csv")
data1
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols #ordinary least squares
```



*   what ols do - it will try to establish a linear relation ship b/w the 2 atributes that given to it



```python
#we are trying to create a model

model=ols('weight~group',data=data1).fit()
#this model will create a relationship b/w weight and group .
```

```python
anova_table=sm.stats.anova_lm(model,typ=1)#type=1 means 1 way anova
#we only want to check one factor that is we only checking the impact of treatment methode
```

```python
anova_table
```

```python
if anova_table.iloc[0,-1]<0.05:#0 th row il -1(last value negative index)
  print('Reject null hypothesis')
else:
  print('Accept null hypothesis')
```

**Two way ANOVA**

```python
data2=pd.read_csv("/content/Anova2.csv")
data2
```



*  Here, 2 facors are present - genotype and years





so we want to know the impact of genotype and years in plant growth.we want its seperate impact and combined impact

so here we want to create a linear relationship b/w genotype with growth,year with growth and genotype+year withgrowth

we want to change the structure of our data set to create linera relation ship.using fn melt of pandas

```python
data2_melt=pd.melt(data2,id_vars=['Genotype'],value_vars=['1_year','2_year','3_year'])
```

```python
data2_melt
```

```python
data2_melt.columns=['Genotype','Years','growth_Value']#change variable to years
```

```python
data2_melt.head()
```

Now we can create the relation ships.

```python
model=ols("growth_Value ~ C(Genotype)+C(Years)+C(Genotype):C(Years)",data=data2_melt).fit()
```

C means categorical(year and genotype)

```python
ano2_table=sm.stats.anova_lm(model,typ=2)
ano2_table
```

```python
if ano2_table.iloc[0,-1]<0.05:
  print('Reject null hypothesis')
if ano2_table.iloc[1,-1]<0.05:
  print('Reject null hypothesis')
if ano2_table.iloc[2,-1]<0.05:
  print('Reject null hypothesis')
else:
  print('Accept null hypothesis')
```

# **STATISTICAL TESTS**

```python
data_new=np.random.randint(20,40,10)
data_new
```



*   H0 :Averge weight of 10 year old children is less than 32 kg
*   H1 :Averge weight of 10 year old children is more than 32 kg



sample size=10.and only one sample is there so we use 1 sample t test


```python
from scipy.stats import ttest_1samp
```

```python
t_stat,p_value=ttest_1samp(data_new,32)
print('t_stat:',t_stat)
print('p_value:',p_value)
```

```python
if p_value<0.05:
  print('Reject null hypothesis')
else:
  print('Accept null hypothesis')
```

**iris data set**



*  We are going to compare the petal length of setosa and versicolor and we want to check whether there is any significance





*   H0 : there is no significant difference b/w these two
*   H1 : there is a significant difference b/w these two



we use both z test and t test.
it is a 2 sample test

```python
iris=pd.read_excel("/content/iris.xls")
iris
```

```python
setosa=iris[iris['Classification']=='Iris-setosa']
versicolor=iris[iris['Classification']=='Iris-versicolor']
```

```python
setosa.head()
```



*   Extract petal lengths for setosa



```python
setosa_PL=setosa['PL']
```

```python
versicolor.head()
```



*  Extract petal length for versicolor



```python
versicolor_PL=versicolor['PL']
```

```python
from scipy.stats import ttest_ind
```

```python
t_stat,p_value=ttest_ind(setosa_PL,versicolor_PL)
print('t_stat:',t_stat)
print('p_value:',p_value)
```

```python
if p_value<0.05:
  print('Reject null hypothesis')
else:
  print('Accept null hypothesis')
```

**new case**

```python
df1=pd.DataFrame({"online shopping preference":('electronics','clothing','books')*10
                  ,'choosen payment methode':('credit card','debit card','paypal')*10})
```

```python
df1
```



*  We want to know whether there is a significance b/w 	online shopping preference and choosen payment methode



```python
#creating a contingency table
contingency=pd.crosstab(df1['online shopping preference'],df1['choosen payment methode'])
contingency
```

```python
from scipy.stats import chi2_contingency
```

```python
chi_stat,p_value,dof,expected=chi2_contingency(contingency)
print('chi_stat:',chi_stat)
print('p_value:',p_value)
print('dof:',dof)
print('expected:',expected)
```

```python
if p_value<0.05:
  print('Reject null hypothesis')
else:
  print('Accept null hypothesis')
```
