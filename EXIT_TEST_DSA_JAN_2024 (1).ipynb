# **2)OUTLIER CORECTION**

Continuous Numerical Columns

```python
num_cols = ['age','ball_controlling_skills', 'body_reflexes',
                    'body_balance', 'jumping_skills', 'penalties_conversion_rate',
                    'mental_strength', 'goalkeeping_skills', 'defending_skills',
                    'passing_skills', 'dribbling_skills', 'shot_accuracy',
                    'body_strength_stamina', 'max_running_speed', 'behaviour_rating',
                    'matches_played', 'fitness_rating', 'trophies_won',
                     'years_of_experience', 'no_of_disqualifications']
```

```python
for i in num_cols:
  plt.figure()
  sns.boxplot(data[i])
  plt.title(i)
  plt.show()
```



*   outlier correction



```python
# Dictionary to store quartiles and outliers
outliers = {}
```

```python
for col in num_cols:
    if col in data.columns:
        Q1 = data[col].quantile(0.25)
        Q2 = data[col].quantile(0.50)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_limit = Q1 - 1.5 * IQR
        upper_limit = Q3 + 1.5 * IQR
        outlier_values = data[(data[col] < lower_limit) | (data[col] > upper_limit)][col]
        outliers[col] = {
            'Q1': Q1,
            'Q2': Q2,
            'Q3': Q3,
            'Outliers': outlier_values.tolist()
        }

# Print the quartiles and outliers for each column
for col, q in outliers.items():
    print(f"{col}: Q1={q['Q1']}, Q2={q['Q2']}, Q3={q['Q3']}, Outliers={q['Outliers']}")
```



*   using clip fn to remove outliers



```python
for col in num_cols:
    if col in data.columns:
        Q1 = data[col].quantile(0.25)
        Q2 = data[col].quantile(0.50)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_limit = Q1 - 1.5 * IQR
        upper_limit = Q3 + 1.5 * IQR
        outlier_values = data[(data[col] < lower_limit) | (data[col] > upper_limit)][col]
        outliers[col] = {
            'Q1': Q1,
            'Q2': Q2,
            'Q3': Q3,
            'Outliers': outlier_values.tolist()
        }
        data[col] = data[col].clip(lower=lower_limit, upper=upper_limit)

# Print the quartiles and outliers for each column
for col, q in outliers.items():
    print(f"{col}: Q1={q['Q1']}, Q2={q['Q2']}, Q3={q['Q3']}, Outliers={q['Outliers']}")
```
