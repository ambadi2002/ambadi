

*   Line chart



A line plot is a type of chart used to display information as a **series of data points** ("markers") connected by straight lines. It is one of the most common forms of data visualization and is typically used to show trends over a continuous interval or time period.

```python
plt.plot(df['YearsAtCompany'],df['PerformanceScore'],color='green',label='line plot',linestyle='--',marker='o')
plt.xlabel('YearsAtCompany')
plt.ylabel('PerformanceScore')
plt.title('Line Plot')
plt.legend()
plt.show()
```

```python
plt.plot(df['YearsAtCompany'],df['PerformanceScore'],label='line plot',color='black',linestyle='--',marker='o')
plt.xlabel("YearsAtCompany")
plt.ylabel("PerformanceScore")
plt.title("Line Plot")
plt.legend()
plt.show()
```

```python
plt.plot(df['YearsAtCompany'],df['PerformanceScore'],label='line1',color='red',linestyle='--',marker='o')
plt.plot(df['YearsAtCompany'],df['Salary'],label='line2',color='yellow',linestyle='--',marker='o')
plt.show
```

Sub Plot

```python
#set up the sub plot with 2 rows and 2 columns
fig,axes=plt.subplots(2,2,figsize=(12,10))

#plot:1 histogram of Ages
axes[0,0].hist(df['Age'])
axes[0,0].set_title('Histogram of Ages')
axes[0,0].set_xlabel('Age')
axes[0,0].set_ylabel('Frequency')

#plot 2: Bar plot of department
axes[0,1].bar(df['Department'],df['Salary'],color='red')
axes[0,1].set_title('Bar Plot of Department')
axes[0,1].set_xlabel('Department')
axes[0,1].set_ylabel('Salary')

#plot 3:pie chart for gender
gender_counts=df['Gender'].value_counts()
axes[1,0].pie(gender_counts,labels=gender_counts.index,autopct='%1.0f%%',startangle=90,colors=['red','blue'])
axes[1,0].set_title('Gender Distribution')

#plot 4: line chart for performence score
axes[1,1].plot(df['YearsAtCompany'],df['PerformanceScore'],color='green',label='line plot',linestyle='--',marker='o')
axes[1,1].set_xlabel('YearsAtCompany')
axes[1,1].set_ylabel('PerformanceScore')
axes[1,1].set_title('Line Plot')
axes[1,1].legend()

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
