```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/supermarket_data.csv')
df.head()
```

```python
df.info()
```

```python
df.columns
```

**LINE CHART**

```python
plt.plot(df['Month'],df['Sales ($)'],label='line plot',color='green',marker='*',linestyle='--')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Line Chart')
plt.legend()
plt.show()
```

```python
plt.plot(df['Month'],df['Customers'],label='line plot',color='red',marker='.',linestyle='-')
plt.xlabel('Month')
plt.ylabel('Customers')
plt.title('Line Chart')
plt.legend()
plt.show()
```

BAR CHART

```python
plt.bar(df['Category'],df['Sales ($)'],label='bar plot',color='blue')
plt.xlabel('Category')
plt.ylabel('Sales ($)')
plt.title('Bar Chart')
plt.legend()
```

```python
plt.bar(df['Region'],df['Revenue ($)'],label='bar plot',color='red')
plt.xlabel('Region')
plt.ylabel('Revenue ($)')
```

Horizondal Barplot

```python
plt.barh(df['Month'],df['Quantity'],label='horizondal bar',color='black')
plt.xlabel('Quantity')
plt.ylabel('Month')
plt.title('Horizondal Bar Chart')
plt.legend()
```

Scatter plot

```python
plt.scatter(df['Sales ($)'],df['Discount (%)'],color='yellow')
plt.xlabel('Sales ($)')
plt.ylabel('Discount (%)')
plt.title('Scatter Plot')
```

Histogram

```python
plt.hist(df['Category'])
```

```python
plt.hist(df['Product'])
```

Pie Chart

```python
plt.pie(df['Sales ($)'],labels=df['Month'],autopct='%1.0f%%')
```

```python
plt.pie(df['Discount (%)'],labels=df['Product'],autopct='%1.0f%%')
```

Sub plot

```python
fig,axes=plt.subplots(2,3,figsize=(18,16))

#line plot
axes[0,0].plot(df['Month'],df['Sales ($)'],label='line plot',color='green',marker='*',linestyle='--')
axes[0,0].set_xlabel('Month')
axes[0,0].set_ylabel('Sales')
axes[0,0].set_title('Line Chart')
axes[0,0].legend()

#Bar chart
axes[0,1].bar(df['Category'],df['Sales ($)'],label='bar plot',color='blue')
axes[0,1].set_xlabel('Category')
axes[0,1].set_ylabel('Sales ($)')
axes[0,1].set_title('Bar Chart')
axes[0,1].legend()

#horizondal bar
axes[0,2].barh(df['Category'],df['Discount (%)'],label='horizondal bar',color='black')
axes[0,2].set_title('Horizondal Bar Chart')
axes[0,2].set_xlabel('Discount (%)')
axes[0,2].set_ylabel('Category')
axes[0,2].legend()

#Scatter plot

axes[1,0].scatter(df['Sales ($)'],df['Discount (%)'],color='yellow')
axes[1,0].set_xlabel('Sales ($)')
axes[1,0].set_ylabel('Discount (%)')
axes[1,0].set_title('Scatter Plot')

#Histogram
axes[1,1].hist(df['Category'])
axes[1,1].set_xlabel('Category')
axes[1,1].set_ylabel('Frequency')
axes[1,1].set_title('Histogram')

#pie chart
axes[1,2].pie(df['Sales ($)'],labels=df['Month'],autopct='%1.0f%%')
axes[1,2].set_title('Pie Chart')
```
